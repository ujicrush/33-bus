```python
import numpy as np
import matplotlib.pyplot as plt
```

### multi

```python
res_path = 'res/multi'

obj_2nd_m = np.load(f"{res_path}/obj_2nd.npy")
lambda_2nd_m = np.load(f"{res_path}/lambda_2nd.npy")
mu_2nd_m = np.load(f"{res_path}/mu_2nd.npy")
previous_rating_m = np.load(f"{res_path}/previous_rating.npy")
previous_cap_m = np.load(f"{res_path}/previous_cap.npy")
ESS_loc_m = np.load(f"{res_path}/ESS_loc.npy")
alpha_store_m = np.load(f"{res_path}/alpha_store.npy")
upperB_save_m = np.load(f"{res_path}/upperB_save.npy")
lowerB_save_m = np.load(f"{res_path}/lowerB_save.npy")
invest_save_m = np.load(f"{res_path}/invest_save.npy")
Elapsed_time_m = np.load(f"{res_path}/Elapsed_time.npy")
total_time_m = np.load(f"{res_path}/total_time.npy", allow_pickle=True)

print("obj_2nd shape:", obj_2nd_m.shape)
print("lambda_2nd shape:", lambda_2nd_m.shape)
print("mu_2nd shape:", mu_2nd_m.shape)
print("previous_rating shape:", previous_rating_m.shape)
print("previous_cap shape:", previous_cap_m.shape)
print("ESS_loc shape:", ESS_loc_m.shape)
print("alpha_store shape:", alpha_store_m.shape)
print("upperB_save shape:", upperB_save_m.shape)
print("lowerB_save shape:", lowerB_save_m.shape)
print("invest_save shape:", invest_save_m.shape)
print("Elapsed_time shape:", Elapsed_time_m.shape)
print("total_time shape:", total_time_m.shape)
```

```python
iterations_m = np.arange(len(upperB_save_m))

plt.figure(figsize=(5, 3))
plt.plot(iterations_m, upperB_save_m, marker='o', label='Upper Bound', linewidth=2)
plt.plot(iterations_m, lowerB_save_m, marker='s', label='Lower Bound', linewidth=2)
plt.title("Convergence of Upper and Lower Bounds", fontsize=12)
plt.xlabel("Iteration", fontsize=10)
plt.ylabel("Convergence Process", fontsize=10)
plt.xticks(iterations_m) 
plt.legend(fontsize=8)
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```

```python
plt.figure(figsize=(5, 3))
plt.plot(iterations_m, Elapsed_time_m, marker='o', linewidth=2)
plt.title("Elapsed time for each iteration", fontsize=12)
plt.xlabel("Iteration", fontsize=10)
plt.ylabel("Time(s)", fontsize=10)
plt.xticks(iterations_m) 
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```
