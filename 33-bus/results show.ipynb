```python
import numpy as np
import matplotlib.pyplot as plt
```

### multi & single

```python
res_path = 'res/multi&single'

obj_2nd_ms = np.load(f"{res_path}/obj_2nd.npy")
lambda_2nd_ms = np.load(f"{res_path}/lambda_2nd.npy")
mu_2nd_ms = np.load(f"{res_path}/mu_2nd.npy")
previous_rating_ms = np.load(f"{res_path}/previous_rating.npy")
previous_cap_ms = np.load(f"{res_path}/previous_cap.npy")
ESS_loc_ms = np.load(f"{res_path}/ESS_loc.npy")
alpha_store_ms = np.load(f"{res_path}/alpha_store.npy")
upperB_save_ms = np.load(f"{res_path}/upperB_save.npy")
lowerB_save_ms = np.load(f"{res_path}/lowerB_save.npy")
invest_save_ms = np.load(f"{res_path}/invest_save.npy")
Elapsed_time_ms = np.load(f"{res_path}/Elapsed_time.npy")
total_time_ms = np.load(f"{res_path}/total_time.npy", allow_pickle=True)

print("obj_2nd shape:", obj_2nd_ms.shape)
print("lambda_2nd shape:", lambda_2nd_ms.shape)
print("mu_2nd shape:", mu_2nd_ms.shape)
print("previous_rating shape:", previous_rating_ms.shape)
print("previous_cap shape:", previous_cap_ms.shape)
print("ESS_loc shape:", ESS_loc_ms.shape)
print("alpha_store shape:", alpha_store_ms.shape)
print("upperB_save shape:", upperB_save_ms.shape)
print("lowerB_save shape:", lowerB_save_ms.shape)
print("invest_save shape:", invest_save_ms.shape)
print("Elapsed_time shape:", Elapsed_time_ms.shape)
print("total_time shape:", total_time_ms.shape)
```

```python
iterations_ms = np.arange(len(upperB_save_ms))

plt.figure(figsize=(5, 3))
plt.plot(iterations_ms, upperB_save_ms, marker='o', label='Upper Bound', linewidth=2)
plt.plot(iterations_ms, lowerB_save_ms, marker='s', label='Lower Bound', linewidth=2)
plt.title("Convergence of Upper and Lower Bounds", fontsize=12)
plt.xlabel("Iteration", fontsize=10)
plt.ylabel("Convergence Process", fontsize=10)
plt.xticks(iterations_ms) 
plt.legend(fontsize=8)
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```

```python
plt.figure(figsize=(5, 3))
plt.plot(iterations_ms, Elapsed_time_ms, marker='o', linewidth=2)
plt.title("Elapsed time for each iteration", fontsize=12)
plt.xlabel("Iteration", fontsize=10)
plt.ylabel("Time(s)", fontsize=10)
plt.xticks(iterations_ms) 
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```

```python
total_time_ms
```
